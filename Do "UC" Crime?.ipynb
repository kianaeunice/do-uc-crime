# Do "UC" Crime?: A Study About Crime within the University of California Campuses

During my time in Berkeley, many of my friends and peers would obsess over claiming the number 1 spot on U.S. News's Top Public University list. Whether Cal would be on the number 1 spot or not, most of us hold a lot of pride towards being a part of an amazing university. Although, many of us also could not ignore what skeletons our school holds.

![alt text](data/images/oski.jpg)

## Background
First, I had to figure out what skeletons I wanted to find. Numerous ideas popped into my mind. *Did I want to find scandals? Did I want to focus on history-based events?* As I was planning my research, YouTube started playing a BuzzFeed Unsolved video, a true-crime based show that focuses on unsolved crimes throughout history. This sparked the idea. I wanted to know what type of crimes happened in the last five years at Cal.

![alt text](data/images/buzzfeedunsolved.gif)

## Acquiring Data
After scouring the web, I thought the best place to find data was on the fbi.gov website, where they shared yearly reports of specific things. I specifically used the Uniform Crime Reporting (UCR) publications from the years of 2014 to 2015. I ended up finding tables including reported crimes in all California colleges and universities. For the purpose of this project, I will start with UC Berkeley.

## Overlook over Cal Crime Statistics

Seeing the plotted graphs gave me an opportunity to look over the different reported crimes that has been reported over the span of these five years. While some fairly high crimes decreased over the years, some actually increased over time.

For instance, motor vehicle theft was at an all time low in 2015, but increased almost triple by 2019. Arson occuring on campus was also at an all time low in 2015, but increased almost double two years later. 


Another thing to look at is what category of crime is mostly reported. In the span of five years between 2014 to 2019, it seems that there is a lot more property crime.

```python
cal_vc_total = sum(berkeley["Violent Crime"])
cal_pc_total = sum(berkeley["Property Crime"])
cal = [cal_vc_total, cal_pc_total]
plt.pie(cal, labels=["Violent Crime", "Property Crime"], autopct='%1.1f%%')
fig = plt.gcf()
fig.set_size_inches(15,15);
```
